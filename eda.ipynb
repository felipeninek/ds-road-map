```python
#import libs
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

#set figsize
sns.set(rc={'figure.figsize':(15, 8)})
```

```python
# load dataset
df = pd.read_csv('data/ensino_estadual.csv')
df.head()
```

```python
df.shape
```

```python
df.info()
```

```python
#DistribuiÃ§Ã£o de frequÃªncia
sns.displot(df['ANOS INICIAIS'])
plt.show()
```

```python
df_escolas = df[df['ANOS INICIAIS'] > 0]

sns.displot(df_escolas['ANOS INICIAIS'], bins=10)
plt.show()
```

```python
#tendÃªncia central e dispersÃ£o
df['ANOS INICIAIS'].describe()
```

```python
#posiÃ§Ã£o relativa
sns.boxplot(y=df_escolas['ANOS INICIAIS'])
plt.show()
```

```python
#AssociaÃ§Ã£o
df_escolas = df[(df['ANOS FINAIS'] > 0) & (df['ANOS INICIAIS'] > 0)]

sns.scatterplot(x=df_escolas['ANOS INICIAIS'], y=df_escolas['ANOS FINAIS'], data=df_escolas)
plt.show()
```
